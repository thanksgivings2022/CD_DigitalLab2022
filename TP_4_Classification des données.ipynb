## 1. K-means, GMM et clustering agglomératif pour le clustering : données synthétiques

Dans cette partie de TP vous allez appliquer les méthodes du clustering indiquées aux jeux de données synthétiques ci-dessous.

```python
from sklearn.datasets.samples_generator import make_blobs

### 4 ordinary round clusters
X1, y1_true = make_blobs(n_samples=400, centers=4,
                       cluster_std=0.50, random_state=0)
X1 = X1[:, ::-1] # flip axes for better plotting

plt.scatter(X1[:, 0], X1[:, 1], s=40, cmap='viridis');

### True clusters visualization
# plt.scatter(X1[:, 0], X1[:, 1], c=y1_true, s=40, cmap='viridis');
```

```python
### 4 ordinary elliptical clusters
rng = np.random.RandomState(13)
X2 = np.dot(X1, rng.randn(2, 2))
y2_true = y1_true 
plt.scatter(X2[:, 0], X2[:, 1], s=40, cmap='viridis');
```

```python
### 2 clusters with different gaussian distributions
np.random.seed(0)
n_samples = 500
X3_1 = 2.0*np.random.randn(n_samples, 2) + np.array([5, 3])
y3_1 = np.zeros(n_samples)
C = np.array([[0., -0.5], [3.5, .7]])
X3_2 = np.dot(np.random.randn(n_samples, 2), C)
y3_2 = np.ones(n_samples)

X3 = np.vstack([X3_1, X3_2])
y3_true = np.vstack([y3_1, y3_2])

plt.scatter(X3[:, 0], X3[:, 1], s=40, cmap='viridis');
```

```python
from sklearn.datasets import make_moons

### 2 MOON cluster
X4, y4_true = make_moons(200, noise=.05, random_state=0)
plt.scatter(X4[:, 0], X4[:, 1], s=40, cmap='viridis');
```
